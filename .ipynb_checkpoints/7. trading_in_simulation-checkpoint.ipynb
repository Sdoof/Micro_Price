```python
def simulation(ite, init=['1','0','0',100,100]):
    initial = init
    res = [initial]
    tick = 0.015625
    current = initial

    for i in range(ite):
        state_in = ''.join(current[:3])
        total_prob = m4.loc[state_in,:].sum()
        random_N = np.random.uniform(0,total_prob)
        state_out = m4.loc[state_in][m4.loc[state_in].cumsum() > random_N].index[0]
        price_move = state_out[3:]
        state_out = state_out[:3]
        if price_move == '00':
            current = list(state_out) + [current[3]] + [current[4]]
        elif price_move == '10':
            current = list(state_out) + [current[3]+tick] + [current[4]]
        elif price_move == '-10':
            current = list(state_out) + [current[3]-tick] + [current[4]]
        elif price_move == '01':
            current = list(state_out) + [current[3]] + [current[4]+tick]
        elif price_move == '0-1':
            current = list(state_out) + [current[3]] + [current[4]-tick]  
        else:
            raise ValueError("Wrong price movement")
        res.append(current)
    res = pd.DataFrame(res, columns = ['price_relation','obi_1','obi_2','price_1','price_2'])
    return res
```

```python
class trader1:
    
    def __init__(self, cash,pnl, inventory,price1,price2,buy_state,sell_state):
        self.cash = cash
        self.inventory = inventory
        self.price1= price1
        self.price2=price2
        self.buy_state=buy_state
        self.sell_state=sell_state
        self.pnl=pnl
        self.pnl_list=[]
        self.inventory_list=[]
        self.tick = 0.015625
        
    def get_pnl(self):
        return self.pnl_list
    
    def get_inventory(self):
        return self.inventory_list
        
    def buy(self,price1,price2):
        self.price1=price1
        self.price2=price2
        self.inventory=self.inventory+1
        self.cash=self.cash-price1+price2-self.tick
        
    def sell(self,price1,price2):
        self.price1=price1
        self.price2=price2
        self.inventory=self.inventory-1
        self.cash=self.cash+price1-price2-self.tick 
        
    def check_buy_sell(self,state,inventory,price1,price2):
        if (state==self.sell_state) and (inventory>=0): 
            self.sell(price1,price2)
        elif (state==self.buy_state) and (inventory<=0): 
            self.buy(price1,price2)
        else:
            self.inventory=inventory #print('nothing')
            
        self.pnl=self.cash+self.inventory*(price1-price2)
        self.pnl_list.append(self.pnl)
        self.inventory_list.append(self.inventory)
        return self.inventory
```

```python
def simulation_with_trader1(ite,buy_state,sell_state, init=['1','0','0',100,100]):
    initial = init
    res = [initial]
    tick = 0.015625
    current = initial
    cash=0
    inventory=0
    cash_list=[]
    inventory_list=[]
    pnl_list=[]
    #initialize trader
    t1=trader1(0,0,0,100,100,buy_state,sell_state)
    
    for i in range(ite):
        state_in = ''.join(current[:3])
        # check if trader does something
        inventory=t1.check_buy_sell(state_in,inventory,current[3],current[4])

        total_prob = m4.loc[state_in,:].sum()
        random_N = np.random.uniform(0,total_prob)
        state_out = m4.loc[state_in][m4.loc[state_in].cumsum() > random_N].index[0]
        price_move = state_out[3:]
        state_out = state_out[:3]
        if price_move == '00':
            current = list(state_out) + [current[3]] + [current[4]]
        elif price_move == '10':
            current = list(state_out) + [current[3]+tick] + [current[4]]
        elif price_move == '-10':
            current = list(state_out) + [current[3]-tick] + [current[4]]
        elif price_move == '01':
            current = list(state_out) + [current[3]] + [current[4]+tick]
        elif price_move == '0-1':
            current = list(state_out) + [current[3]] + [current[4]-tick]  
        else:
            raise ValueError("Wrong price movement")
        res.append(current)
    res = pd.DataFrame(res, columns = ['price_relation','obi_1','obi_2','price_1','price_2'])
    return t1.get_pnl()
```

```python
def matrix_Gstar_BC_G1(M_input=m4,step_forward = 10):
    
    #trainsaction matrix
    tm = M_input.iloc[:,:36]

    s1_up = M_input.iloc[:,36:72] #    am.loc[:,s1_up_index]
    s1_up_mat=np.matrix(s1_up)

    s1_down = M_input.iloc[:,72:108] # am.loc[:,s1_down_index]
    s1_down_mat=np.matrix(s1_down)
    
    s2_up = M_input.iloc[:,108:144] #am.loc[:,s2_up_index]
    s2_up_mat=np.matrix(s2_up)
    
    s2_down =  M_input.iloc[:,144:180] #am.loc[:,s2_down_index]
    s2_down_mat=np.matrix(s2_down)

    Q = np.matrix(tm.to_numpy())
    n = Q.shape[0]
    tick = 0.015625

    R = np.zeros((n, 4))
    R[:,0] = s1_up.sum(axis = 1, skipna = True)
    R[:,1] = s1_down.sum(axis = 1, skipna = True)
    R[:,2] = s2_up.sum(axis = 1, skipna = True)
    R[:,3] = s2_down.sum(axis = 1, skipna = True)
    R = np.matrix(R)

    T=s1_up_mat+s1_down_mat+s2_up_mat+s2_down_mat
    K = np.matrix([[tick,0], [-tick,0],[0,tick],[0,-tick]])
    G1 = np.linalg.inv(np.identity(n) -Q).dot(R).dot(K)
    B = np.linalg.inv(np.identity(n) -Q).dot(T)
    T_series =  list(map(np.matrix,[s1_up,s1_down,s2_up,s2_down]))
     
    Gstar = G1
    BC = B

    for i in range(step_forward-1):
        Gstar = Gstar+BC.dot(G1)
        BC = BC.dot(B)
    Gstar = pd.DataFrame(Gstar,index = tm.index,columns = ['price1_change_'+str(step_forward)+'step','price2_change_'+str(step_forward)+'step'])
    
    return Gstar,BC,G1,B,Q,T,R,K
```

```python
Gstar,BC,G1,B,Q,T,R,K=matrix_Gstar_BC_G1(M_input=m4,step_forward=100)
```
